### Predicting Regression

#### Predicting Bike Rentals

```python
df_bikes = df_bikes.drop(['casual', 'registered'], axis=1)
```

```python
# Saving data for future use
df_bikes.to_csv('bikes_rentals_cleaned.csv', index=False)
```

### Declaring Predictors & Target columns

```python
X = df_bikes.iloc[:, :-1]
y = df_bikes.iloc[:, -1]
```

```python
# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
# Silencing warnings
warnings.filterwarnings('ignore')
```

### Modeling Linear Regression

```python
# Initialize the M.L model
lin_reg = LinearRegression()
```

```python
# Fitting on training set
lin_reg.fit(X_train, y_train)
```

```python
# Observing Predictions
y_pred = lin_reg.predict(X_test)
```

```python
# Basic comparison with Root Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
```

```python
# Squared Root
rmse = np.sqrt(mse)
```

```python
# Outcomes
print("RMSE: %0.2f" % (rmse))
```

```python
# Checking on Rental expected per day
df_bikes['cnt'].describe()
```

## XGBoost

```python
# # Initialize the XGBRegressor model
xg_reg = XGBRegressor()
```

```python
# Fitting XGBRegressor on training set
xg_reg.fit(X_train, y_train)
```

```python
# Observing Predictions
y_pred = xg_reg.predict(X_test)
```

```python
# Basic comparison with Root Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
# Squared Root
rmse = np.sqrt(mse)
```

```python
# XGBRegressor outcome
print("RMSE: %0.2f" % (rmse))
```

#### Cross-Validation

```python
# Initilize the model
model = LinearRegression()
```

```python
# Implement cross-validation on the model
scores = cross_val_score(model, X, y, scoring ='neg_mean_squared_error', cv=10)
```

```python
# Checking the RMSE
rmse = np.sqrt(-scores)
```

```python
# Checking results
print("Reg rmse: ", np.round(rmse, 2))
```

```python
print("RMSE mean: %0.2f" % (rmse.mean()))
```

#### Cross-Validation with XGBoost

```python
# Initilize the XGB model
model = XGBRegressor()
```

```python
# Implement cross-validation on the model
scores = cross_val_score(model, X, y, scoring ='neg_mean_squared_error', cv=10)
```

```python
# Checking the XGBRegressor RMSE
rmse = np.sqrt(-scores)
```

```python
# Checking results
print("Reg rmse: ", np.round(rmse, 2))
```

```python
print("RMSE mean: %0.2f" % (rmse.mean()))
```

#### Reloading with 'header' None

```python
df_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)
```

```python
df_census.head()
```

```python
# Directly fixing column names
df_census.columns=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship',
                  'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
```

```python
df_census.head()
```

#### Null-Values

```python
df_census.info()
```

```python
df_census = df_census.drop(['education'], axis=1)
```

```python
df_census = pd.get_dummies(df_census)
df_census.head()
```

#### Target & Predictor Columns

```python
# Deleting one predictor column
df_census = df_census.drop('income_ <=50K', axis=1)
```

```python
X = df_census.iloc[:, :-1]
y = df_census.iloc[:,-1]
```

### Logistic Regression

```python
# Setting a cross-validation function
def cross_val(classifier, num_splits=10):
    model = classifier
    scores = cross_val_score(model, X, y, cv=num_splits)
    print('Accuracy:', np.round(scores, 2))
    print('Accuracy mean: %0.2f' % (scores.mean()))
```

```python
cross_val(LogisticRegression())
```

#### The XGBoost Classifier

```python
cross_val(XGBClassifier(n_estimators=5))
```
